```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import CIFAR10
import torch.optim as optim
```

```python
import matplotlib.pyplot as plt
```

```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
transform=transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))    
])
```

```python
dataset = ImageFolder(root="./data/PetImages", transform=transform)
```

```python
dataset.classes
```

```python
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
trainset, testset = random_split(dataset, [train_size, test_size])
```

```python
trainLoader = DataLoader(trainset, batch_size=64, shuffle=True, num_workers=4, pin_memory=True)
testLoader = DataLoader(testset, batch_size=64, shuffle=False, num_workers=4, pin_memory=True)
```

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv_layers=nn.Sequential(
            
            nn.Conv2d(in_channels=3,out_channels=8,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,stride=2),

            nn.Conv2d(in_channels=8,out_channels=16,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,stride=2),

            nn.Conv2d(in_channels=16,out_channels=32,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,stride=2),

            nn.Conv2d(in_channels=32,out_channels=64,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,stride=2)
            
        )

        self.fc_layers=nn.Sequential(
            nn.Linear(8*8*64,256),
            nn.ReLU(),

            nn.Linear(256,64),
            nn.ReLU(),

            nn.Linear(64,2)
        )
    def forward(self,x):
        x= self.conv_layers(x)
        x=x.view(x.size(0),-1)
        x= self.fc_layers(x)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model=CNN().to(device)
```

```python
criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters())
```

```python
epochs=10
train_loss=[]
val_loss=[]
best_val_loss=float("inf")
for epoch in range(epochs):
    
    running_loss=0.0
    model.train()
    for xb,yb in trainLoader:
        xb = xb.to(device)  
        yb = yb.to(device)  
        optimizer.zero_grad()
        output=model(xb)
        loss=criterion(output,yb)
        loss.backward()
        optimizer.step()
        running_loss+=loss.item()


    avg_train=running_loss/len(trainLoader)
    train_loss.append(avg_train)

    
    running_loss=0.0
    model.eval()
    with torch.no_grad():
        for xb,yb in testLoader:
            xb = xb.to(device)  
            yb = yb.to(device)  
            output=model(xb)
            loss=criterion(output,yb)
            running_loss+=loss.item()


    avg_test=running_loss/len(testLoader)
    val_loss.append(avg_test)


    print(f"{epoch+1}/{epochs}, training_loss:{avg_train}, testing_loss:{avg_test}")

    if(avg_test<best_val_loss):
        best_val_loss=avg_test
        torch.save(model.state_dict(),"best_model.pt") #.pt or .pth
```

```python
model.load_state_dict(torch.load("best_model.pt"))
```

```python
plt.figure(figsize=(16,10))
plt.plot(train_loss)
plt.plot(val_loss)
```

```python


correct_labels=0
total_labels=0
model.eval()
with torch.no_grad():
    for xb,yb in testLoader:
        xb = xb.to(device)  
        yb = yb.to(device)  
        output=model(xb)
        _,predicted=torch.max(output,1)
        correct_labels+= (predicted==yb).sum().item()
        total_labels+=yb.size(0)

print(f"Accuracy={correct_labels/total_labels*100}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# 1. Set model to evaluation mode
model.eval()

all_preds = []
all_labels = []

# 2. Collect predictions and true labels from the test loader
with torch.no_grad():
    for xb, yb in testLoader:
        xb = xb.to(device)
        outputs = model(xb)
        _, preds = torch.max(outputs, dim=1)
        
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(yb.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)

# 3. Compute metrics
accuracy = accuracy_score(all_labels, all_preds)
precision = precision_score(all_labels, all_preds)
recall = recall_score(all_labels, all_preds)
cm = confusion_matrix(all_labels, all_preds)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print("\nConfusion Matrix:")
print(cm)

# 4. Plot Confusion Matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dataset.classes)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
# Install gradio first if you haven't already: !pip install gradio
import gradio as gr
import torchvision.transforms as transforms
from PIL import Image

def predict_image(img):
    if img is None:
        return {}
    
    # Convert to RGB to support RGBA (PNGs with transparency) and Grayscale images
    img = img.convert("RGB")
    
    # Preprocess the PIL image to match our model's input requirements
    eval_transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))    
    ])
    
    # Apply transformations and add a batch dimension (shape: [1, 3, 128, 128])
    img_tensor = eval_transform(img).unsqueeze(0).to(device)
    
    # Set model to evaluation mode and predict
    model.eval()
    with torch.no_grad():
        outputs = model(img_tensor)
        probabilities = torch.softmax(outputs[0], dim=0)
        
    # Return a dictionary of class names and probabilities
    classes = ['Cat', 'Dog']  # More robust than referencing dataset.classes
    return {classes[i]: float(probabilities[i]) for i in range(len(classes))}

# Set up the Gradio interface
interface = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil", label="Upload Cat/Dog Image"),
    outputs=gr.Label(num_top_classes=2, label="Predictions"),
    title="🐱 Cat vs Dog Classifier 🐶",
    description="Upload an image to see if the model predicts a Cat or a Dog with confidence levels.<br>💻 **GitHub Repository:** [SAHIL-AGARWAL-IN/CNN_CAT_DOG](https://github.com/SAHIL-AGARWAL-IN/CNN_CAT_DOG)"
)

# Launch the interface (it will display inline inside the notebook)
interface.launch(share=False)
```
